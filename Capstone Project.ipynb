# <div align="center">3.2 Available Metrics:</div>

In evaluating the performance of the classification models, several metrics are available:

1. Accuracy: The proportion of correctly classified articles out of the total articles.
2. Precision: The proportion of true positive predictions among all positive predictions.
3. Recall: The proportion of true positive predictions among all actual positive instances.
4. F1-Score: The harmonic means of precision and recall, providing a balanced measure of model performance.

# <div align="center">3.3 Metric Selection and Reasoning:</div>

Given the complex nature of news categorization and the importance of balanced performance evaluation, the F1-Score is chosen as the primary metric for assessing the classification models. The F1-Score is a harmonic mean of precision and recall, offering a balanced measure of model performance that accounts for both false positives and false negatives.

In the context of news classification, it's crucial to achieve a balance between accurately identifying articles belonging to a specific category (precision) and capturing all relevant articles within that category (recall). The F1-Score provides a unified metric that considers both aspects, making it well-suited for evaluating the model's effectiveness in accurately categorizing news articles across all categories.

Furthermore, the F1-Score is particularly valuable when dealing with imbalanced datasets, where the number of articles in each category may vary significantly. By incorporating both precision and recall into a single metric, the F1-Score offers a holistic view of the model's performance without being biased towards the majority class.

```python
#Lowercasing text in 'TITLE' column
df['lower'] = df['TITLE'].str.lower()
```

Lowercasing text in 'TITLE' column:

This process involves converting all text in the 'TITLE' column to lowercase letters. Lowercasing ensures uniformity in text data by treating uppercase and lowercase versions of the same word as identical. It simplifies subsequent text processing and analysis tasks.

```python
df.head()
```

```python
REMOVE_PUNCT = string.punctuation

def remove_punctuation(text):
    """custom function to remove the punctuation"""
    return text.translate(str.maketrans('', '', REMOVE_PUNCT ))

df["punc_removed"] = df["lower"].apply(lambda text: remove_punctuation(text))


```

Removing Punctuation:

Punctuation marks such as periods, commas, exclamation points, and quotation marks are removed from the text in the 'TITLE' column. Punctuation removal is essential for text analysis tasks like natural language processing (NLP) and classification, where punctuation may not contribute to the semantic meaning of the text.

```python
df.head()
```

```python
#Removing StopWords
STOPWORDS = set(stopwords.words('english'))
def remove_stopwords(text):
    return " ".join([word for word in str(text).split() if word not in STOPWORDS])
df["stopwords_removed"] = df["punc_removed"].apply(lambda text: remove_stopwords(text))
```

Removing StopWords:

Stopwords are common words that often appear frequently in text data but typically do not contribute much to the overall meaning or context of the text. Examples of stopwords include "the," "is," "are," "and," etc. Removing stopwords helps reduce noise in the text data and focuses the analysis on more meaningful words.

```python
df.head()
```

```python
lemmatizer = WordNetLemmatizer()

# Define a function to lemmatize words in a text
def lemmatize_words(text):
    return " ".join([lemmatizer.lemmatize(word) for word in text.split()])

# Apply lemmatization to the "stopwords_removed" column and create a new column "lemmatized_stopwords"
df["lemmatized_stopwords"] = df["stopwords_removed"].apply(lambda text: lemmatize_words(text))
```

Lemmatizing (since Stemming may create non-existent/incorrect words):

Lemmatization is the process of reducing words to their base or root form, known as the lemma. Unlike stemming, which simply chops off the ends of words to remove affixes, lemmatization considers the word's context and morphological analysis to produce valid lemmas. Lemmatization ensures that the resulting words are valid and meaningful, preventing the creation of non-existent or incorrect words that may arise from stemming. It helps maintain the integrity of the text data during preprocessing for tasks such as sentiment analysis or classification.

# Results using CountVectorizer


```python
# convert data into vectors
vectorizer = CountVectorizer()
x = vectorizer.fit_transform(df['lemmatized_stopwords'])
y = df['CATEGORY']

x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42,stratify=df.CATEGORY)
print('Training Data ',x_train.shape,y_train.shape)
print('Test Data     ',x_test.shape,y_test.shape)

```


This code snippet converts textual data from the 'lemmatized_stopwords' column into numerical vectors using CountVectorizer, a method for converting text data into a matrix of token counts. The 'fit_transform' function fits the vectorizer to the text data and transforms it into a sparse matrix representation where each row corresponds to a document and each column represents a unique word in the corpus. The resulting matrix 'x' contains the vectorized representation of the text data. The data is split into training and testing sets using train_test_split, with 80% of the data used for training ('x_train', 'y_train') and 20% for testing ('x_test', 'y_test'). The shapes of the training and testing datasets are printed to confirm the dimensions of the data

```python
results = pd.DataFrame(columns=['Model', 'Accuracy', 'Precision', 'Recall', 'F1-score'])

models_name = ['Logistic Regression', 'Multinomial NaiveBayes']

model_list = [LogisticRegression(), MultinomialNB()]

for idx, model in enumerate(model_list):
    clf = model.fit(x_train, y_train)
    predictions = model.predict(x_test)
    
    accuracy = accuracy_score(y_test, predictions)
    precision = precision_score(y_test, predictions, average='weighted')
    recall = recall_score(y_test, predictions, average='weighted')
    f1 = f1_score(y_test, predictions, average='weighted')
    
    results.loc[idx] = [models_name[idx], accuracy, precision, recall, f1]

results
```

This code initializes an empty DataFrame named 'results' with columns for model names, accuracy, and F1-score. It defines a list of model names and a corresponding list of model objects for logistic regression and multinomial Naive Bayes classifiers. It then iterates through each model in the list, trains the model on the training data (x_train, y_train), makes predictions on the test data (x_test), and calculates accuracy and F1-score using sklearn's accuracy_score and f1_score functions. The results are stored in the 'results' DataFrame and sorted based on accuracy in descending order. Finally, the sorted results DataFrame is printed. the naive bayes refers to the assumption of conditional independence among the features.
